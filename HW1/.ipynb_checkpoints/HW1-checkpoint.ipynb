```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn import datasets

circles_data, circles_data_labels = datasets.make_circles(n_samples=100, factor=0.1, noise=0.1)
```

```python
# Convert data to tensors
X = torch.tensor(circles_data, dtype=torch.float32)
y = torch.tensor(circles_data_labels, dtype=torch.long)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
class NeuralNetwork(nn.Module):
    def __init__(self):
        super(NeuralNetwork,self).__init__()
        self.fc1=nn.Linear(2,64)
        self.fc2=nn.Linear(64,32)
        self.fc3=nn.Linear(32,2)
        
    def forward(self,x):
        x=torch.relu(self.fc1(x))
        x=torch.relu(self.fc2(x))
        x=self.fc3(x)
        return x
```

```python
model = NeuralNetwork()

# Define the loss function and the optimizer
criterion = nn.CrossEntropyLoss()  # Change loss function to CrossEntropyLoss
optimizer = optim.Adam(model.parameters(), lr=0.01)
```

```python
epochs = 100
for epoch in range(epochs):
    optimizer.zero_grad()
    outputs = model(x_train)
    loss = criterion(outputs, y_train)
    loss.backward()
    optimizer.step()
    if epoch % 10 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')
        
        # Calculate accuracy
        with torch.no_grad():
            predicted = torch.argmax(outputs, dim=1)
            correct = (predicted == y_train).sum().item()
            accuracy = correct / len(y_train)
            print(f'Accuracy: {accuracy:.4f}')
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# 定義繪製決策邊界的函數
def plot_decision_boundary(X, y, model):
    # 取得特徵值的最小和最大值，用於確定繪圖範圍
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    # 生成網格點
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
    
    # Debugging: 檢查 xx 和 yy 的shape
    print("xx shape:", xx.shape)
    print("yy shape:", yy.shape)
    # 使用模型預測網格點的類別
    Z = model(torch.tensor(np.c_[xx.ravel(), yy.ravel()], dtype=torch.float32)).detach().numpy()
    
    print("Z shape:", Z.shape)
    
    # 將概率轉換為類別（0或1）
    Z = np.argmax(Z, axis=1)
    Z = Z.reshape(xx.shape)
    print("Z shape:", Z.shape)
    # 定義淺色背景色塊來表示不同類別的特徵空間
    cmap_light = ListedColormap(['orange', 'yellow'])
    # 繪製背景色塊
    plt.contourf(xx, yy, Z, cmap=cmap_light)
    # 繪製數據點
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, s=50)

# 繪製數據點和決策邊界
plt.figure(figsize=(8, 6))
plot_decision_boundary(circles_data, circles_data_labels, model)
plt.title("circles_data_labels")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
